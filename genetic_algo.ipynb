# Генетический алгоритм для рюкзака

## 1. Initializing first population 

```python
def init_population(chromosome_len, population_size, knap_weight, weights):
  random_bits = [[[random.choice([0,1]) for i in range(chromosome_len)], 0, 0] for _ in range(population_size)]
  return random_bits
```

## 2. Fitness function

```python
def make_relevant(chromosome, weights, costs, knap_weight):
  while True:
    current_weigth = 0
    current_cost = 0

    for i in range(len(chromosome[0])):
      if chromosome[0][i]:
        # print(chromosome[0][i], i, weights, weights[i])
        current_weigth += weights[i]
        current_cost += costs[i]

    if current_weigth <= knap_weight:
      return chromosome, current_weigth, current_cost

    
    indices = [i for i, x in enumerate(chromosome[0]) if x]
    chromosome[0][np.random.choice(indices)] = 0
```

```python
def fitness_func(population, knap_weight, weights, costs):
  fitnessed = [] # chromosome, weight, cost

  for chrom in population:
    current_weigth = 0
    current_cost = 0

    for i in range(len(chrom[0])):
      if chrom[0][i]:
        current_weigth += weights[i]
        current_cost += costs[i]

    if current_weigth > knap_weight:
      chrom, current_weigth, current_cost = make_relevant(chrom, weights, costs, knap_weight)
      # print('changed')

    fitnessed.append([chrom[0], current_weigth, current_cost])
    fitnessed.sort(reverse=True, key = lambda x: x[2])
  return fitnessed
```

## 3. Selection

```python
def selection(fitness_population):
  group_number = np.random.choice(4, 1, p=[0.6, 0.2, 0.15, 0.05])[0]
  GROUPS_AMOUNT = 4
  group_size = len(fitness_population) // GROUPS_AMOUNT
  offset = group_size * group_number
  random_group = fitness_population[offset:offset + group_size]
  random_ind = np.random.choice(group_size)
  return random_group[random_ind]
```

## 4. Crossover

```python
def crossover(fitness_population):
  el_a, el_b = selection(fitness_population), selection(fitness_population)
  if el_a[0] == el_b[0]:
    return crossover(fitness_population)

  random_ind = np.random.choice(len(el_a[0]))
  el_a[0], el_b[0] = el_a[0][:random_ind] + el_b[0][random_ind:], el_b[0][:random_ind] + el_a[0][random_ind:]
  return fitness_population
```

## 5. Mutation

```python
def mutation(fitness_population):
  for el in fitness_population:
    for ind in range(len(el[0])):
      to_change = np.random.choice(2, 1, p=[0.9, 0.1])[0]
      if to_change:
        el[0][ind] = int(bool(not el[0][ind]))
  return fitness_population
```

## 6. Start function

```python
def genetic_knapsack(knap_weight, weights, costs, n_iter, population_size = 12):
  generation = 0
  population = init_population(len(weights), population_size, knap_weight, weights)

  current_generation = fitness_func(population, knap_weight, weights, costs)
  current_result = [[], 0, 0]
  while generation < n_iter:
    current_generation = mutation(crossover(current_generation))
    current_generation = fitness_func(current_generation, knap_weight, weights, costs)
    if current_result[2] < current_generation[0][2]:
      if current_generation[0][1] == knap_weight:
        return current_generation[0]
      current_result = current_generation[0]
    generation += 1

  return current_result
```

## 0. Get distance table

```python
def get_distance(p1, p2):
  return sqrt((p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2)
```

```python
def get_distance_table(data):
  distances = [[0 for _ in data] for k in data]

  for ind1, p1 in enumerate(data):
    for ind2, p2 in enumerate(data):
      distances[ind1][ind2] = get_distance(p1, p2)
  return distances
```

```python
def fitness(population, n_vertex, distances):
  fitnessed = []
  for i in range(len(population)):
    current_len = 0
    for j in range(n_vertex - 1):
      current_len += distances[j][j+1]
    current_len += distances[population[i][0][n_vertex - 1]][population[i][0][0]]
    fitnessed.append([population[i][0], current_len])
    fitnessed.sort(key = lambda x: x[1])
  return fitnessed
```

```python
def partition(el, changed):
  while el in changed.keys():
    el = changed[el]
  return el
```

```python
def update_dictionary(dct):
  marked_keys = set()
  while set(dct.keys()) - marked_keys:
    diff_keys = list(set(dct.keys()) - marked_keys)
    key = diff_keys[0]
    value = dct[key]
    keys_to_change = []
    while value in dct.keys():
      keys_to_change.append(key)
      key = value
      value = dct[value]

    for key in keys_to_change:
      dct[key] = value
    marked_keys.union(set(keys_to_change))
    marked_keys.add(key)

  return dct
```

## 5. Mutation

```python
def mutation(fitness_population):
  c = 0
  ind_list = [i for i in range(len(fitness_population))]
  for el_ind, el in enumerate(fitness_population):
    for ind in range(len(el[0])):
      to_change = np.random.choice(2, 1, p=[0.9, 0.1])[0]
      if to_change:
        another_el_ind = np.random.choice(ind_list[:el_ind] + ind_list[el_ind + 1:])
        another_el = fitness_population[another_el_ind]

        another_el_in_el_ind = el[0].index(another_el[0][ind])
        el[0][another_el_in_el_ind],  el[0][ind] = el[0][ind], el[0][another_el_in_el_ind]

        el_ind_in_another_el = another_el[0].index(el[0][another_el_in_el_ind])
        another_el[0][el_ind_in_another_el], another_el[0][ind] = another_el[0][el_ind_in_another_el], another_el[0][ind]
  return fitness_population
```

## 6. Start function

```python
def genetic_traveling_salesman(data, n_iter, population_size = 12):
  generation = 0
  distances = None
  if hasattr(data, 'node_coords') and data.node_coords:
    coords = list(data.node_coords.values())
    distances = get_distance_table(coords)
  else:
    distances = list(data.as_name_dict().get('edge_weights'))
  n_vertex = len(distances)
  population = init_paths(n_vertex, population_size)
  
  # print(population)
  current_generation = fitness(population, n_vertex, distances)
  current_result = [[], 10 ** 9]
  while generation < n_iter:
    current_generation = mutation(crossover(current_generation))
    current_generation = fitness(population, n_vertex, distances)
    if current_result[1] > current_generation[0][1]:
      current_result = current_generation[0]
    generation += 1

  return current_result
```

```python
from time import time
```

```python
def test_genetic_knapsack(dataset):
  knap_weight, weights, costs = dataset['c'][0], dataset['w'], dataset['p']
  start_time = time()
  result = genetic_knapsack(knap_weight, weights, costs, 1000)
  end_time = time() - start_time
  right_answer = sum([dataset['p'][ind] for ind, item in enumerate(dataset['s']) if item])
  delta = abs(result[2] - right_answer)
  return delta, end_time, result
```
